```python
import os
import shutil
import pandas as pd
from tqdm import tqdm
```

# train 데이터

```python
# csv 파일 읽기
df = pd.read_csv("./t_라벨_참외_0정상.csv")
df
```

```python
# 1(열매), 2(꽃), 5(줄기) 이미지 개수 확인 -> 2329개여야 함
use_df = df.loc[df['classID'].isin([1])]
use_df.to_csv("./temp.csv", index=False)
```

```python
# 1(열매), 2(꽃), 5(줄기) 이미지 개수 확인 -> 2329개여야 함
len(df.loc[df['classID'].isin([2]), 'image'])
```

```python
src_path = "source_images_path"
dst_path = "destination_folder_path"

# 복사할 이미지 목록 생성
images_to_copy = df.loc[df['classID'].isin([1]), 'image'].tolist()

# 해당되는 이미지만 복사
for img in tqdm(images_to_copy, desc="Copying images", unit="image"):
    src_file = os.path.join(src_path, img)
    dst_file = os.path.join(dst_path, img)

    # 파일이 존재하는지 확인
    if os.path.exists(src_file):
        shutil.copy(src_file, dst_file)
    else:
        print(f"Warning: File not found - {src_file}")

print("Train 파일 복사 완료")
```
