```python
# 데이터 요약 정보 확인 함수 정의
def resumetable(df):
    print(f'Data Shape: {df.shape}')
    summary = pd.DataFrame(df.dtypes, columns=['데이터 타입'])
    summary = summary.reset_index()
    summary = summary.rename(columns={'index': 'Feature'})
    summary['데이터 수'] = df.count().values
    summary['결측치'] = df.isnull().sum().values
    summary['고유값'] = df.nunique().values
    summary['최소값'] = df.min().values
    summary['최대값'] = df.max().values

    summary['First Value'] = df.iloc[0].values
    summary['Second Value'] = df.iloc[1].values

    return summary

resumetable(data)
```

```python
# 상관계수 히트맵
plt.figure(figsize=(20, 15))
sns.heatmap(data.iloc[:, 1:].corr(), annot=True, fmt='.2f', cmap='RdBu')    # cmap: 색상
plt.show()
```

```python
# 결혼 성사 여부 비율 확인
married_count = data['married'].value_counts()
print(married_count)
print('-----'*10)
print(f'Not married: {married_count[0] / married_count.sum()*100:.2f}%  |  Married: {married_count[1] / married_count.sum()*100:.2f}%')

# 막대 그래프
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
married_count.plot(kind='bar', color=['skyblue', 'pink'])

# 파이 차트
plt.subplot(1, 2, 2)
married_count.plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'pink'])
plt.show()
```
